```python
# Strip any leading or trailing spaces from the column names
df1.columns = df1.columns.str.strip()
```

```python
#drop columns with too many missing values = population 
df1.drop(['Population'], axis=1, inplace=True) 
```

```python
#using stratergy 1 
df1['Adult Mortality'].fillna(df1['Adult Mortality'].median(), inplace=True)
df1['Polio'].fillna(df1['Polio'].median(), inplace=True)
df1['Diphtheria'].fillna(df1['Diphtheria'].median(), inplace=True)
df1['Hepatitis B'].fillna(df1['Hepatitis B'].median(), inplace=True)
df1['GDP'].fillna(df1['GDP'].median(), inplace=True)
# df1['Population'].fillna(df1['Population'].median(), inplace=True)
df1['Income composition of resources'].fillna(df1['Income composition of resources'].median(), inplace=True)
df1['Schooling'].fillna(df1['Schooling'].median(), inplace=True)
df1['Total expenditure'].fillna(df1['Total expenditure'].median(), inplace=True)
df1['thinness  1-19 years'].fillna(df1['thinness  1-19 years'].median(), inplace=True)
df1['thinness 5-9 years'].fillna(df1['thinness 5-9 years'].median(), inplace=True)
```

```python
#using stratergy 2 - use the mean 
df1['Alcohol'].fillna(df1['Alcohol'].mean(), inplace=True)
df1['BMI'].fillna(df1['BMI'].mean(), inplace=True)
```

```python
# Capping outliers using the IQR method
Q1 = df1[numerical_columns].quantile(0.25)
Q3 = df1[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

df1_capped = df1.copy()

for col in numerical_columns:
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]
    df1_capped[col] = df1_capped[col].clip(lower=lower_bound, upper=upper_bound)

# Check the capped data
df1_capped.describe()

df1.isnull().sum()
```
